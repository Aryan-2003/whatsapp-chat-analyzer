```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Notice group CSE 2020-24.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
# converrt messages_date type
df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date':'date'},inplace=True)
df.head()
```

```python
df.shape
```

```python
# seperate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop('user_message',axis=1,inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute

df.head()
```

```python
words = []
for message in df['message']:
    words.extend(message.split())
```

```python
len(words)
```

```python
from urlextract import URLExtract
extractor = URLExtract()
urls = extractor.find_urls("Hey you can go to https://www.ggogle.com and search for https://facebook.com")
urls
```
